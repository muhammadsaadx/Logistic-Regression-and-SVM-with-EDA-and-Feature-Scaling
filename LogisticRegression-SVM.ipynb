```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data
np.random.seed(42)
data = {
    'A': np.random.randint(1, 100, 50),
    'B': np.random.randn(50) * 10 + 50,
    'C': np.random.choice(['X', 'Y', 'Z'], 50)
}

# Create DataFrame
df = pd.DataFrame(data)
print(df.head())

# Basic statistics
print(df.describe())

# Visualization
plt.figure(figsize=(10, 5))
sns.scatterplot(x='A', y='B', hue='C', data=df, palette='viridis')
plt.title('Scatter Plot of A vs B')
plt.show()
```
